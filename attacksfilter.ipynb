```python

attacks=["BENIGN", "Bot", "DDoS", "DoS GoldenEye", "DoS Hulk", "DoS Slowhttptest", "DoS slowloris", "FTP-Patator", "Heartbleed", "Infiltration", "PortScan", "SSH-Patator", "Web Attack Â– Brute Force", "Web Attack  Sql Injection", "Web Attack  XSS"]


benign=2359289

dict_attack={
"Bot":1966,
"DDoS":41835,
"DoS GoldenEye":10293,
"DoS Hulk":231073,
"DoS Slowhttptest":5499,
"DoS slowloris":5796,
"FTP-Patator":7938,
"Heartbleed":11,
"Infiltration":36,
"PortScan":158930,
"SSH-Patator":5897,
"Web Attack  Brute Force":1507,
"Web Attack  XSS":652,
"Web Attack  Sql Injection":21}
```

```python
for i in dict_attack:
    a=0
    b=0
    this = open('./attacks/' + i + '.csv', 'w')
    this.write(str(main_labels) + '\n')
    benign_num=int(benign/(dict_attack[i]*(9/5)))
    with open("./CSVs/all_data.csv", "r") as file:
        while True:
            try:
                line = file.readline()
                line = line[:-1]
                k=line.split(',')
                if k[83]=='BENIGN':
                    rnd=random.randint(1,benign_num)
                    if rnd==1:
                            this.write(str(line)+'\n')
                            b+=1
                if  k[83]==i:
                    this.write(str(line)+'\n')
                    a+=1
                else:
                    continue
            except: 
                break
    this.close()
    print(i ,"file is completed\n attack:%d\n benign:%d\n\n\n " %(a,b))
```
